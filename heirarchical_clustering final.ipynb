Hierarchical cluster analysis 


Aim : To make clusters using the given data set of customer spending score and annual income

```python
# importing important packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#reading the file

dataset = pd.read_csv("Mall_Customers.csv")
x=dataset.iloc[ : , : ].values
```

```python
# importing scipy.cluster.hierarchy to draw dendrograms
#A dendrogram is a diagram that shows the hierarchical relationship between objects. 
import scipy.cluster.hierarchy as sch
```

#dendrogram using ward (method) 1





```python
dendrogram = sch.dendrogram(sch.linkage(x, method = "ward"))
plt.title("Dendrogram")
plt.xlabel("customers")
plt.ylabel("Euclidean distance")
plt.show()
```

```python
#dendrogram1 = sch.dendrogram(sch.linkage(x, method = "average"))
```

first cluster object for making 5 clusters

```python
#agglomerative clustering
# agglomerative clustering is a type of hierarchical clustering to cluster 
# a group of objects or data points into one
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=5)
y_hc=clustering.fit_predict(x)
```

Scatter plot for first cluster object 


```python
#scatter plots
plt.scatter(x[y_hc==0,0],x[y_hc==0,1], c="red", label="cluster 1")
plt.scatter(x[y_hc==1,0],x[y_hc==1,1], c="blue", label="cluster2")
plt.scatter(x[y_hc==2,0],x[y_hc==2,1], c="green", label="cluster 3")
plt.scatter(x[y_hc==3,0],x[y_hc==3,1], c="orange", label="cluster 4")
plt.scatter(x[y_hc==4,0],x[y_hc==4,1], c="violet", label="cluster 5")

plt.title("heirarchy clustering")
plt.xlabel("annual income(k$)")
plt.ylabel("spending score(1-100)")
plt.legend()
plt.show()
```
